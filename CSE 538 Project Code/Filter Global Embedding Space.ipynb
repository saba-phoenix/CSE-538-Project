```python
import io
import numpy as np

def load_vec(emb_path, nmax=50000):
    vectors = []
    word2id = {}
    with io.open(emb_path, 'r', encoding='utf-8', newline='\n', errors='ignore') as f:
        next(f)
        for i, line in enumerate(f):
            word, vect = line.rstrip().split(' ', 1)
            vect = np.fromstring(vect, sep=' ')
            assert word not in word2id, 'word found twice'
            vectors.append(vect)
            word2id[word] = len(word2id)
            if len(word2id) == nmax:
                break
    id2word = {v: k for k, v in word2id.items()}
    embeddings = np.vstack(vectors)
    return embeddings, id2word, word2id
```

#### Load all the book's word embedding, find common words from the both Global and Book embedding and save the vector in two seperate file. It will speed up the embedding alignment Process

```python
import os
import pandas as pd
from collections import Counter

PATH = '..\\Emotion Flow GitHub\\data_all\\'

for root, dirs, files in os.walk(PATH):
    for i,name in enumerate(files):
        if name.endswith((".vec")):
            if 'filtered' in name: 
                continue
            genere = (root.split("\\"))[3]
            outcome = (root.split("\\"))[4]
            print(genere, outcome, name)
            target_embed,init_id2word,init_word2id = load_vec(root+'\\'+name)
            
            f = open(root+'\\'+name.replace('.vec','')+'_filtered.vec', 'w')
            
            count = 0
            for w in init_word2id.keys():
                if w in temp_word2id:
                    count = count + 1
        
            f.write(str(count))
            f.write(" 300")
            f.write('\n')

            for w in init_word2id.keys():
                if w in temp_word2id:
                    x = temp_word2id[w]
                    vec = source_embed[x]
                    f.write(w)
                    f.write(" ")
                    for v in vec:
                        f.write(str(v))
                        f.write(" ")
                    f.write('\n')
            f.close()
            #break
```
